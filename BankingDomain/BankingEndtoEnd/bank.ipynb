```python
from sqlalchemy import create_engine
import pandas as pd

engine = create_engine("mysql+mysqlconnector://root:password@localhost/bank")

query = "SELECT * FROM customer"
df = pd.read_sql(query, engine)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Estimated Income'].min()
```

```python
bins=[1,100000,300000,float('inf')]
labels=['Low','Medium','High']

df['Income band']=pd.cut(df['Estimated Income'],bins=bins,labels=labels,right=False)
```

```python
df['Income band'].value_counts()
```

```python
#examine the distrubutiion of unique categories in categorical columns
categorical_cols=df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income band']]

for col in categorical_cols:
    print(f"value counts for '{col}':")
    display(df[col].value_counts())
```

```python
for i,predictor in enumerate(df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income band']]):
    plt.figure(i)
    sns.countplot(data=df,x=predictor,hue='Nationality')
```

```python
#histplot of value counts for different occupations

for col in categorical_cols:
    if col == 'Occupation':
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title("Histogram of Occupation count")
    plt.xlabel(col)
    plt.ylabel("count")
    plt.show()
```

Numerical analysis

```python
numerical_cols=['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Foreign Currency Account']

plt.figure(figsize=(15,10))
for i,col in enumerate(numerical_cols):
    plt.subplot(3,2,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
    plt.show()
```

#Heatmap

```python
numerical_cols=['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Foreign Currency Account']

correlation_matrix=df[numerical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix,annot=True,cmap='crest',fmt=".2f")
plt.show()
```
